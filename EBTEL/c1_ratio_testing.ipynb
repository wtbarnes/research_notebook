# $c_1$ Ratio Testing
In EBTEL, there are three coefficients derived from static equilibrium loop solutions:
$$
c_1=\frac{\mathcal{R}_{TR}}{\mathcal{R}_C},\\
c_2=\frac{\bar{T}}{T_a},\\
c_3=\frac{T_0}{T_a}.
$$
The last two coefficients are reasonably well approximated by $c_2=0.9$ and $c_3=0.6$. The first coefficient is the most dynamic and hardest to formulate.

## EBTEL-1
In the first EBTEL paper [Klimchuk et al. (2008)](http://adsabs.harvard.edu/abs/2008ApJ...682.1351K), $c_1$ was determined to be approximately constant at $c_1=4$ from static equilibrium solutions. 

## EBTEL-2
In the second EBTEL paper [Cargill et al. (2012a)](http://adsabs.harvard.edu/abs/2012ApJ...752..161C), the calculation of $c_1$ was revisited because $c_1=4$ did not give correct solutions for short, ($L\sim25$ Mm) hot ($T>3$ MK) loops. A new "baseline" value of $c_1$ was calculated as $c_1=2$ from the work of [Martens (2010)](http://adsabs.harvard.edu/abs/2010ApJ...714.1290M).

## Variable $c_1$
Additional physics can be included in the calculation of $c_1$ by corrections to the baseline value.

### 1. Gravity & Radiation
We can rewrite the ratio, correcting only for gravity, as,
$$
c_1=\frac{\mathcal{R}_{TR}}{\mathcal{R}_C}=\left(\frac{\mathcal{R}_{TR}}{\mathcal{R}_{TR}(g=0)}\right)\left(\frac{\mathcal{R}_{TR}(g=0)}{\mathcal{R}_C(g=0)}\right)\left(\frac{\mathcal{R}_{C}(g=0)}{\mathcal{R}_C}\right).
$$

The first ratio is unity. The second is the nominal baseline value, $c_{1,b}$. The third ratio can be rewritten,
$$
\frac{\mathcal{R}_{C}(g=0)}{\mathcal{R}_C} \approx \frac{(n^2\Lambda(\bar{T}))_{g=0}}{n^2\Lambda(\bar{T})} = \frac{(n^2\chi\bar{T}^{\alpha})_{g=0}}{n^2\chi\bar{T}^{\alpha}}\approx\frac{(\bar{n}^2)_{g=0}}{\bar{n}^2} = \frac{p_0^2}{\bar{p}^2}
$$

Noting that $p(s)\approx p_0\exp{-((2L/\pi\bar{\lambda})\sin{(\pi s/2L)})}$ and that $\bar{p}\approx p(s=2L/5)$, the total graviational correction can be written as,
$$
c_1 = c_{1,b}\exp\left(-\frac{4L}{\pi\bar{\lambda}}\sin{\left(\frac{\pi}{5}\right)}\right)
$$

Furthermore, a correction can be applied (in the same manner as the gravitational correction) to account for additional parts of the radiative loss function. Including this correction, the whole $c_1$ coefficient can be written as,
$$
c_1 = c_{1,b}\left(\frac{\mathcal{R}_C(g=0,\alpha=-2/3)}{\mathcal{R}_C(g=0,\bar{T})}\right)\exp\left(-\frac{4L}{\pi\bar{\lambda}}\sin{\left(\frac{\pi}{5}\right)}\right)
$$

### 2. Phases of Evolution
For equilibrium conditions, loops are well-described by $T\sim n^{1/2}$. During the phase of loop evolution where a loop is in or near equilibrium, $c_{1,b}\approx2$.

When $T\sim n^{1/2}$ does not hold, such a baseline value is not appropriate (e.g. during radiative and enthalpy-driven cooling where $T\sim n^2$ for short loops and $T\sim n$ for longer loops). During such conditions, 
$$
\frac{1}{\gamma-1}\frac{d\bar{p}}{dt} = -\frac{\mathcal{R}_C}{L}(1+c_1), \\
\frac{d\bar{n}}{dt} = -\frac{c_2c_1}{c_3}\frac{(\gamma-1)\mathcal{R}_C}{2k_B\bar{T}L\gamma},
$$

Generalizing $T\sim n^{\delta}$, we note that $p\sim kn^{\delta+1}$ such that $dp/dt\sim(\delta+1)kn^{\delta}(dn/dt)$. Equating and solving for $c_1$,
$$
\frac{\frac{1}{p}\frac{dp}{dt}}{\frac{1}{n}\frac{dn}{dt}} = \delta + 1 = \gamma\frac{c_3}{c_2}\frac{(1+c_1)}{c_1} \\
c_1 = \left(\frac{c_2(1+\delta)}{\gamma c_3} + 1\right)^{-1}
$$

For long loops, $\delta=1$ and $c_1=1.25$ while for short loops, $\delta=2$ and $c_1=0.6$.

### Total
From above, we have two baseline values: $c_{1,b,eqm}=2$ and $c_{1,b,rad}=0.6$. The graviational and radiation corrections remain unchanged. To define a smooth transition between the two values, we use the ratio of the density $\bar{n}$ to the equilibrium density $n_{eq}$ such that,
$$
c_1 = c_{1,b,eqm}c_{1,grav}c_{1,rad},\quad n<n_{eq}, \\
c_1 = \left(\frac{2c_{1,b,eqm} + c_{1,b,rad}((n/n_{eq})^2 - 1)}{1+(n/n_{eq})^2}\right)c_{1,grav}c_{1,rad},\quad n>n_{eq}
$$

### Conductive Phase Correction
To correct for the overestimate of the density in EBTEL, we will apply a conductive value of $c_1$, $c_{1,cond}$ (in the same way that we use a radiative value of $c_1$, $c_{1,rad}$) and we fix $c_{1,eqm}=2$. Thus, our new prescription will be, dropping the loss and gravity corrections,
$$
c_1 = \left(\frac{2c_{1,eqm} + c_{1,cond}((n_{eq}/n)^2 - 1)}{1+(n_{eq}/n)^2}\right),\quad n<n_{eq}, \\
c_1 = \left(\frac{2c_{1,eqm} + c_{1,rad}((n/n_{eq})^2 - 1)}{1+(n/n_{eq})^2}\right),\quad n>n_{eq}
$$

Make a quick-plot function for easy viewing.

```python
def quick_plot(ebtel_data_file,hydrad_instance,two_fluid=False,fig_size=(8,8)):
    """Plot EBTEL/HYDRAD comparison easily"""
    data = np.loadtxt(ebtel_data_file)
    
    fig,ax = plt.subplots(2,2,figsize=fig_size,sharex=True)
    if not two_fluid:
        t,Te,n,c1 = data[:,0],data[:,1],data[:,2],data[:,-3]
        Ti = Te
    else:
        t,Te,Ti,n,c1 = data[:,0],data[:,1],data[:,2],data[:,3],data[:,-3]
    n_hydrad_interp = np.interp(t,hydrad_instance.time,hydrad_instance.timeseries['ne'])
    T_hydrad_interp = np.interp(t,hydrad_instance.time,hydrad_instance.timeseries['Te'])
    err_n = (n - n_hydrad_interp)/n_hydrad_interp
    err_T = (Te - T_hydrad_interp)/T_hydrad_interp
    
    ax[0,0].plot(t,Te/1e+6,color=sns.color_palette('deep')[0],label=r'$\mathrm{EBTEL}$')
    ax[0,0].plot(t,Ti/1e+6,color=sns.color_palette('deep')[0],linestyle='dashed')
    ax[0,0].plot(hydrad_instance.time,hydrad_instance.timeseries['Te']/1e+6,color=sns.color_palette('deep')[2],label=r'$\mathrm{HYDRAD}$')
    ax[0,0].plot(hydrad_instance.time,hydrad_instance.timeseries['Ti']/1e+6,color=sns.color_palette('deep')[2],linestyle='dashed')
    ax[0,1].plot(t,n/1e+8,color=sns.color_palette('deep')[0])
    ax[0,1].plot(hydrad_instance.time,hydrad_instance.timeseries['ne']/1e+8,color=sns.color_palette('deep')[2])
    ax[1,0].plot(t,c1,color=sns.color_palette('deep')[0])
    ax[1,1].plot(t,err_n,color=sns.color_palette('deep')[0],label=r'$\Delta n/n$')
    ax[1,1].plot(t,err_T,color=sns.color_palette('deep')[2],label=r'$\Delta T_e/T_e$')
    
    ax[0,0].set_xscale('log')
    ax[0,0].set_xlim([t[0],t[-1]])
    ax[0,0].set_ylim([0,10])
    ax[0,1].set_ylim([0,50])
    ax[1,0].set_ylim([.5,10])
    ax[1,1].set_ylim([-.5,.5])
    
    ax[1,0].set_xlabel(r'$t$ $\mathrm{(s)}$')
    ax[1,1].set_xlabel(r'$t$ $\mathrm{(s)}$')
    ax[0,0].set_ylabel(r'$T$ $\mathrm{(MK)}$')
    ax[0,1].set_ylabel(r'$n$ $(10^8\,\mathrm{cm}^{-3})$')
    ax[1,0].set_ylabel(r'$c_1$')
    ax[1,1].set_ylabel(r'$\epsilon$')
    
    ax[0,0].legend(loc=1)
    ax[1,1].legend(loc=4)
    
    plt.tight_layout()
    plt.show()
```

## Motivation for Corrections to Baseline Values
In EBTEL-2, this new value $c_{1,b,eqm}\approx2$ is derived from the scaling laws of [Martens (2010)](http://adsabs.harvard.edu/abs/2010ApJ...714.1290M). Can we justify a value of $c_{1,b,eqm}=4$?

Using the approach of Martens (2010), the energy equation can be recast as,
$$
\epsilon\frac{d^2\eta}{ds^2}=\eta^{\mu}-\xi,\quad\mu=-\frac{2(2-\alpha)}{7},\quad\xi=\frac{7}{3+2\alpha}, \\
\epsilon = \frac{2}{1-2(2-\alpha)/7}\left(\frac{2(2-\alpha)}{7}\right)^2\frac{1}{B^2(\lambda+1,1/2)}, \\
\lambda = \frac{3/2+\alpha}{2(2-\alpha)},
$$

where $\eta=(T/T_a)^{7/2}$. The equation can be solved for $u=\eta^{-\mu}$ as
$$
\frac{s}{L} = \beta_r(u,\lambda+1,1/2).
$$

By finding the point at which conduction switches from a loss to a gain, it can be shown that 
$$
\eta_0 = \left(\frac{7}{3+2\alpha}\right)^{-\frac{7}{2(2-\alpha)}}
$$

The dimensionless coronal radiative losses can be written
$$
\mathscr{R}_C = \int^1_{s/L(\eta_0)}\mathrm{d}(s/L)\,\eta^{\mu} = \xi(1-s/L(\eta_0)) - \epsilon\left(\frac{d\eta}{ds}\right)_{s/L=s/L(\eta_0)}
$$

Similarly, the dimensionless TR losses can be written,
$$
\mathscr{R}_{TR} = \int^{s/L(\eta_0)}_0\mathrm{d}(s/L)\,\eta^{\mu} = \xi s/L(\eta_0) + \epsilon\left(\frac{d\eta}{ds}\right)_{s/L=s/L(\eta_0)}
$$

Noting that
$$
\left(\frac{d\eta}{ds}\right)_{s/L=s/L(\eta_0)} = \sqrt{\frac{2}{\epsilon}\left(\frac{\eta_0^{\mu+1}}{\mu+1} - \xi\eta_0\right)}
$$

and definining
$$
\phi = \xi\beta_r(\eta_0^{-\mu},\lambda+1,1/2) + \sqrt{2\epsilon\left(\frac{\eta_0^{\mu+1}}{\mu+1} - \xi\eta_0\right)},
$$

the $c_1$ EBTEL coefficient can be written as 
$$
c_1 = \frac{\mathscr{R}_{TR}}{\mathscr{R}_{C}} = \frac{\phi}{\xi - \phi}
$$

Now, let's define some functions and calculate it.

```python
def mu(alpha):
    return -2.*(2.-alpha)/7.

def xi(alpha):
    return 7./(3.+2.*alpha)

def llambda(alpha):
    return (3./2. + alpha)/(2.*(2.-alpha))

def eta0(alpha):
    power = -7./(2.*(2.-alpha))
    return (7./(3.+2.*alpha))**power
    
def epsilon(alpha):
    term1 = 2./(1.-2.*(2.-alpha)/7.)
    term2 = 2.*(2.-alpha)/7.
    l = llambda(alpha)
    return term1*(term2**2)/(sc_special.beta(l+1,1./2.)**2.)

def phi(alpha):
    x = xi(alpha)
    l = llambda(alpha)
    eps = epsilon(alpha)
    n0 = eta0(alpha)
    m = mu(alpha)
    term1 = x*sc_special.betainc(l+1.,1./2.,n0**(-m))
    term2 = 2.*eps*(n0**(m+1.)/(m+1.) - x*n0)
    return term1 + term2**(1./2.)

def analytic_c1(alpha):
    return phi(alpha)/(xi(alpha) - phi(alpha))
```

We can check our calculation using the results from the EBTEL-2 paper. In Appendix B, they claim,
$$
c_1(\alpha=-1/2) = 1.76, \\
c_1(\alpha=-2/3) = 2.095.
$$

```python
print('c1(alpha=%f) = %f'%(-1./2.,analytic_c1(-1./2.)))
```

```python
print('c1(alpha=%f) = %f'%(-2./3.,analytic_c1(-2./3.)))
```

So we are able to reproduce the results of Cargill et al. (2012) with these expressions. We note from the results above that near the peak density, we are cooling through temperatures around 4 MK. According to the power-law loss function in Klimchuk et al. (2008) (i.e. EBTEL-1),
$$
\Lambda(T) = 3.53\times10^{-13}T^{-3/2},\quad 1.5\times10^6<T<3.5\times10^6 \\

So while the approximate power-law scalings $\alpha=-1/2,-2/3$ do give $c_1\approx2$, the form of the power-law radiative loss function more appropriate to the temperature regime during the transition from conductive cooling to radiative and enthalpy-driven cooling (where we are claiming $c_1=2$) is more well-described by $\alpha\approx-1\sim-3/2$ such that $c_1\approx4$ is a better estimate of the ratio of the transition region to coronal losses.

```python
def powerlaw_alpha(T):
    alpha = [1.9,-1,0,-5/4,1/3,-1,1/2]
    log_tbins = [0.,4.97,5.67,6.18,6.55,6.90,7.63,300.]
    for i in range(len(alpha)):
        if np.log10(T) > log_tbins[i] and np.log10(T) <= log_tbins[i+1]:
            return alpha[i],log_tbins[i],log_tbins[i+1]
        
    raise ValueError("Couldn't find temperature in bins.")
```

## Determining $c_1$ with the Transition Region Crossing Time

Instead of setting $c_1=4$ as a constant during the conductive cooling period, we could instead choose to decrease $c_1$ from $c_1^{(a)}$ to $c_1^{(b)}$ over a timescale $\tau_{c_1}$. 

We express this timescale as the transition region crossing time such that
$$
\tau_{c_1} = \frac{\ell}{C_s}=\frac{\ell}{(\gamma p/\rho)^{1/2}},
$$
where $\ell=\varepsilon L$ is the transition region thickness and $C_s$ is the sound speed. Furthermore, we may choose to define the timescale by an integer multiple $n$ of crossing times such that 
$$
\tau_{c_1} = n\varepsilon L\left(\frac{\rho}{\gamma p}\right)^{1/2}.
$$

Thus, our full implementation of $c_{1,b,eqm}$ can be written as,
$$
c_{1,b,eqm} = \begin{cases} 
      \frac{c_1^{(a)}(1-(t/\tau_{c_1})^2) + 2c_1^{(b)}(t/\tau_{c_1})^2)}{1+(t/\tau_{c_1})^2}, & 0\leq t\leq \tau_{c_1} \\
      c_1^{(b)}, & t>\tau_{c_1}  
   \end{cases}
$$

Note that this is fine for a single heating event that starts at $t=0$. For multiple heating events (or events that start at $t>0$), we need to "zero" our absolute time according to the start time of the most recent event. 

An axample of such a function for $c_1^{(a)}=4,\,\,c_1^{(b)}=2$ and a timescale of $\tau_{c_1}=1500$ s is shown below.

```python
c0 = 2
c1 = 4
a = 1.0/np.log(c0/c1)
tau_c = 4500
t = np.linspace(0,5000,1000)

def dynamic_c1(time):
    c1_list = []
    for t in time:
        if t<tau_c:
            c1_list.append((c0*(1-(t/tau_c)**2) + 2*c1*(t/tau_c)**2)/(1+(t/tau_c)**2))
            #c1_list.append(c0*np.exp(-t/(a*tau_c)))
        else:
            c1_list.append(c1)
        
    return np.array(c1_list)
```

```python
plt.figure(figsize=(8,4))
plt.plot(t,dynamic_c1(t))
plt.ylim([1,5])
plt.xlabel(r'$t$')
plt.ylabel(r'$c_1$')
plt.axhline(2,color='black',linestyle='--')
plt.axhline(4,color='black',linestyle='--')
plt.axvline(tau_c,color='red',linestyle=':')
```

The timescale $\tau_{c_1}$ seems to be quite small for the cases we are looking at. Simplifying our above expressions a bit more,
$$
\tau_{c_1}=\frac{n\varepsilon L}{\sqrt{\gamma p_0/\rho}}=\frac{n\varepsilon L}{\sqrt{\gamma2k_Bc_3\bar{T}/c_2m_p}}=n\varepsilon L\sqrt{\frac{m_pc_2}{2\gamma k_Bc_3}}\bar{T}^{-1/2}.
$$
Noting that $c_2\sim0.9,\,\,c_3\sim0.6$ and $\gamma=5/3$, this can be rewritten as
$$
\tau_{c_1}\approx n\varepsilon L\frac{3}{2}\sqrt{\frac{m_p}{5k_B\bar{T}}}.
$$

So we can plot the timescale as a function of $\bar{T}$. Using $n=2$ and $\varepsilon\sim0.2$ for a loop half-length of $L=40 Mm$,

```python
n=2
eps_tr = 0.2
L_half = 40e+8
def tau_c1_estimate(temp):
    return 1.5*n*eps_tr*L_half*(ac.m_p.cgs.value/(5.*ac.k_B.cgs.value*temp))**(0.5)
t_average = np.logspace(6,7,1000)
plt.plot(t_average,tau_c1_estimate(t_average))
plt.xscale('log')
```
