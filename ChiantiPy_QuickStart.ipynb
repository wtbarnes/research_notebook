<h1>ChiantiPy: The Python interface to the CHIANTI atomic database for astrophysical spectroscopy</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Quick Start Tutorial</h2>

<p>First import the ChiantiPy routines that allow us to access the CHIANTI database.</p>

```python
import chianti.core as ch
```

<p>What we will really be interested in are various properties of the Fe XIV emissivities as a function of tmeperature and density. Define an array of temperatures (using numpy).

```python
t=10.**(5.8 + 0.05*np.arange(21.))
```

<p>In ChiantiPy, temperatures are typically given in Kelvin and densities as the number electron density per cubic cm. Then, construct fe14 as would be typically done.</p>

```python
fe14 = ch.ion('fe_14',temperature=t,eDensity=1.e+9)
```

<h3>Level populations</h3>

<p>Now, plot the population of the top 10 levels of Fe XIV as a function of temperature</p>

```python
fe14.popPlot()
```

<h3>Emissivities</h3>

<p>Plot the emissivities for the top (default=10) lines for Fe XIV in the specified wavelength range. Since there are 21 temperatures involved, a singe temperature is selected.</p>

```python
fe14.emissPlot(wvlRange=[200.,220.])
```

<p>Now try plotting emissivities for a specific temperature, t[2]=7.9e+5.</p>

```python
fe14.emissPlot(index=2,wvlRange=[210.,220.])
```

<p>Now, let's create a table of emission information over a specified wavelength range for a given temperature.</p>

```python
fe14.emissList(wvlRange=[200,220],index=10,relative=1)
```

<p>Important note: an output file could also be created by setting the keyword outFile to the name of the desired output file.</p>

<h3>G(n,T) function</h3>

<p>Now we would like to plot the radiative loss function G(n,t) for some chosen wavelength. The following command prompts the user to make a choice of wavelength. Then, the specified emissivities are plotted (for the top 3 wavelengths in the range) followed by a plot of G(n,t) versus temperature at the chosen wavelength.</p>

```python
fe14.gofnt(wvlRange=[210.,220.],top=3)
```

<h3>Intensity Ratios</h3>

<p>The following command first prompts the user to select the numerator and denominator of the intensity ratio. Then, it produces a plot of the emissivity for all wavelengths in wvlRange followed by a plot of the intensity ratio versus temperature.</p>

```python
fe14.intensityRatio(wvlRange=[210.,225.])
```

<p>Now let's say that when we created our Fe XIV ion, we fixed the temperature and chose to vary the density.</p>

```python
d = 10.**(6 + 0.1*np.arange(61))
fe14 = ch.ion('fe_14', temperature = 2.e+6, eDensity = d)
```

```python
fe14.intensityRatio(wvlRange=[210.,225.])
```

<p>To obtain ratios of lines widely separated in wavelength, the wvlRanges keyword can be used.</p>

```python
fe12 = ch.ion('fe_12',temperature=t,eDensity=1.e+9)
fe12.intensityRatio(wvlRanges=[[190.,200.],[1240.,1250.]])
```

<h3>Spectra of a single ion</h3>

<p>The following series of commands will calculate the spectrum of a Fe XIV over the specified wavelength range and filter it with the default filter which is a gaussian with a 'resolving power' of 1000 which gives a gaussian width of wvl/1000.</p>

```python
fe14 = ch.ion('fe_14',temperature = 2.e+6, eDensity = 1.e+9)
wvl = 200. + 0.125*np.arange(801)
fe14.spectrum(wvl)
plt.plot(wvl, fe14.Spectrum['intensity'])
```

<p>Let's import some additional filters from ChiantiPy and see how the plot changes.</p>

```python
import chianti.filters as chfilters
fe14.spectrum(wvl,filter=(chfilters.gaussian,.04))
```

<p>This calculates the spectrum of Fe XIV for a gaussian filter witha  width of 0.04 Angstroms.</p>

```python
plt.plot(wvl,fe14.Spectrum['intensity'])
```

<h3>Free-free and free-bound continuum</h3>

<p>The module continuum provides the ability to calculate the free-free and free-bound spectrum for a larger number of individual ions.</p>

```python
temperature = 2.e+7
c = ch.continuum('fe_25',temperature=temperature)
wvl = 1. + 0.002*np.arange(4501)
c.freeFree(wvl)
plt.plot(wvl,c.FreeFree['rate'])
c.freeBound(wvl)
plt.plot(wvl,c.FreeBound['rate'])
```

<p>In the continuuum calculations, the specified ion, Fe XXV in this case, is the target ion for the free-free calculation. For the free-bound calculation, the specified ion is also the target ion. In this case, the radiative recombination spectrum of Fe XXV recombining to form Fe XXIV is returned.</p>

<h3>Spectra of multiple ions and continuum</h3>

```python
temperature = [1.e+6,2.e+6]
density=1.e+9
wvl = 200. + 0.05*arange(2001)
emeasure = [1.e+25,1.e+25]
s=ch.spectrum(temperature,density,wvl,filter=(chfilters.gaussian,.2),em=emeasure);
```

```python
subplot(311)
plot(wvl,s.Spectrum['integrated'])
subplot(312)
plot(wvl,s.Spectrum['intensity'][0]*emeasure[0])
ylabel('$erg cm^{-2} s^{-1} sr^{-1} \AA^{-1}$')
subplot(313)
plot(wvl,s.Spectrum['intensity'][1]*emeasure[1])
xlabel('Wavelength ($\AA$)')
```

<p>It is also possible to specify a selection of ions by means of the ionList keyword.</p>

```python
temperature = 2.e+7
density = 1.e+9
wvl = 1. + 0.002*arange(4501)
emeasure = 1.e+25
s = ch.spectrum(temperature, density, wvl, em = emeasure, filter = (chfilters.gaussian,.005));
```

```python
figure(figsize=(8,8))
plot(wvl,s.FreeFree['intensity'])
plot(wvl,s.FreeBound['intensity'])
plot(wvl,s.FreeBound['intensity']+s.FreeFree['intensity'])
xlabel('Wavelength ($\AA$)',fontsize=14)
```

<h3>Radiative loss rate</h3>

<p>The radiative loss rate can be calculated as a function of temperature and density</p>

```python
temp = 10.**(4. + 0.05*arange(81))
rl = ch.radLoss(temp,1.e+4,minAbund=2.e5)
rl.radLossPlot()
```
