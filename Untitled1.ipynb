```python
import os
import tempfile

import numpy as np
import astropy.units as u
from astropy.coordinates import SkyCoord
from sunpy.net import Fido,attrs as a,vso
from sunpy.time import TimeRange
from sunpy.map import Map
import matplotlib.pyplot as plt
import ipyvolume

from synthesizAR.extrapolate import PotentialField,trace_fieldlines,peek_projections

%matplotlib inline
```

```python
# Download data
q = Fido.search(a.Time(TimeRange('2015/08/06 15:19:21', 30*u.s)),
                a.Instrument('HMI'), vso.attrs.Physobs('LOS_magnetic_field'))
```

```python
q
```

```python
with tempfile.TemporaryDirectory() as tmpdirname:
    file_path = Fido.fetch(q, path=os.path.join(tmpdirname, '{file}.fits'))
    magnetogram = Map(file_path)
```

```python
# Rotate
magnetogram = magnetogram.rotate(order=3)
# Crop to AR
c = u.Quantity([-330,-369],'arcsec')
fov = u.Quantity([250,200],'arcsec')
blc = SkyCoord(*(c - fov/2), frame=magnetogram.coordinate_frame)
trc = SkyCoord(*(c + fov/2), frame=magnetogram.coordinate_frame)
magnetogram_ar = magnetogram.submap(blc,trc)
```

```python
# Resample
magnetogram_ar_resampled = magnetogram_ar.resample([150, 150]*u.pixel)
```

```python
# Plot
fig = plt.figure(figsize=(10,20))
ax = fig.add_subplot(311,projection=magnetogram)
magnetogram.plot(title='Full',axes=ax,cmap='hmimag',vmin=-1.5e3,vmax=1.5e3)
magnetogram.draw_rectangle(blc,trc.Tx - blc.Tx, trc.Ty - blc.Ty, axes=ax, color='k', lw=2)
ax = fig.add_subplot(312,projection=magnetogram_ar)
magnetogram_ar.plot(title='Active Region',axes=ax,cmap='hmimag',vmin=-1.5e3,vmax=1.5e3)
ax = fig.add_subplot(313,projection=magnetogram_ar_resampled)
magnetogram_ar_resampled.plot(title='Resampled',axes=ax,cmap='hmimag',vmin=-1.5e3,vmax=1.5e3)
plt.tight_layout()
```

```python
shape_z = 30*u.pixel
width_z = (magnetogram_ar_resampled.scale.axis1 * shape_z).to(u.radian).value * magnetogram_ar_resampled.dsun
```

```python
pf = PotentialField(magnetogram_ar_resampled,width_z,shape_z)
```

```python
B_field = pf.extrapolate()
```

```python
peek_projections(B_field)
```

```python
lines = pf.trace_fieldlines(B_field,100,loop_length_range=[10,50]*u.Mm,
                            mask_threshold=0.1,
                            close_threshold=0.1,
                            direction=1
                           )
```

```python
pf.peek(lines,line_frequency=1)
```
