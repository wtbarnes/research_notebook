```python
import os,sys
from imp import reload
import numpy as np
import matplotlib.pyplot as plt
import seaborn.apionly as sns
sys.path.append('/home/wtb2/Documents/hercules/hercules/')
sys.path.append('/home/wtb2/Documents/EBTEL_analysis/src/')
import quick_look as ql
%matplotlib inline
```

Import a couple of different hydrad datasets and compare how grid options can affect results.

## Single

```python
hydrad_labs=['500_lowRes','500_noCons_linRest','quick','quicker','perfect']
hydrad_res = {}
results_root = '/home/wtb2/Desktop/hp1_hydrad_runs/single/HYDRAD_%s'
```

```python
for hl in hydrad_labs:
    tmp_inst = ql.QuickViewer(os.path.join(results_root%hl,'Results'))
    tmp_inst.load_results()
    hydrad_time = tmp_inst.time
    tmp_inst.make_timeseries(lower_percent=0.45,upper_percent=0.45)
    hydrad_res[hl] = tmp_inst.timeseries
```

Plot the results. Do they differ? How?

```python
fig,ax = plt.subplots(2,1,figsize=(8,8),sharex=True)
for hl,i in zip(hydrad_labs,range(len(hydrad_labs))):
    ax[0].plot(hydrad_time,hydrad_res[hl]['Te']/1e+6,label=hl,color=sns.color_palette('deep')[i],linewidth=2)
    ax[1].plot(hydrad_time,hydrad_res[hl]['ne']/1e+8,label=hl,color=sns.color_palette('deep')[i],linewidth=2)
ax[0].set_xscale('log')
ax[1].set_xscale('log')
ax[0].set_xlim([0,5000])
ax[1].set_xlim([0,5000])
ax[0].legend(loc='best')
```
