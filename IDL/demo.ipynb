## Defining Functions & Procedures

```idl
FUNCTION mysin, x
    return, sin(x)
END
```

```idl
print, mysin(2)
```

```idl
PRO hello_world
    print,"Hello, World!"
END
```

## Inline Plots

```idl
x = 2*!PI*0.01*indgen(100)
y1 = sin(x)
y2 = cos(x)
;loadct,39
;window,xsize=800,ysize=500
;tplot=plot,x,y1,title='Trig Functions',xtitle='Radians',ytitle='Amplitude',xrange=[0,6];,charsize=1.5;,background=255,color=0
;oplot,x,y2,color=50
```

```idl
plot(x,y1,xrange=[0,6],title='Trig Functions',xtitle='Radians',ytitle='Amplitude',color='red')
OPLOT(x,y2)
```
